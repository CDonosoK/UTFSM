```python
import numpy as np
```

```python
# Firma función 1):
'''
input:
numbers       : (ndarray) Numpy Array 1D.

output:
suma          : (double) Suma de los números en el array usando solo la operación suma nativa de python.

'''
def manual_summation(numbers):
    suma = 0
    for num in numbers:
        suma += num
    return suma

# Firma función 2):
'''
input:
numbers       : (ndarray) Numpy Array 1D.

output:
suma          : (double) Suma de los números en el array ordenados de menor a mayor usando solo la operación suma nativa de python.

'''
def ascending_sort_summation(numbers):
    numbers = -np.sort(-numbers)
    suma = 0
    for num in numbers:
        suma += num
    return suma

# Firma función 3):
'''
input:
numbers       : (ndarray) Numpy Array 1D.

output:
suma          : (double) Suma de los números en el array ordenados de mayor a menor usando solo la operación suma nativa de python.

'''
def descending_sort_summation(numbers):
    numbers = np.sort(numbers)
    suma = 0
    for num in numbers:
        suma += num
    return suma

# Firma función 4):
'''
input:
numbers       : (ndarray) Numpy Array 1D.

output:
suma          : (double) Suma de los números en el array usando solo la operación suma nativa de python 
                         y tomando todos los números de forma aleatoria sin repetición en el proceso de sumar.

'''
def random_summation(numbers):
    indexs = np.random.permutation(numbers.size)
    suma = 0
    for index in indexs:
        suma += numbers[index]
    return suma

# Firma función 5):
'''
input:
numbers       : (ndarray) Numpy Array 1D.

output:
suma          : (double) Suma de los números en el array usando np.sum de numpy.

'''
def numpy_summation(numbers):
    suma = np.sum(numbers)
    return suma

# Firma función 6):
'''
input:
numbers       : (ndarray) Numpy Array 1D.
suma          : (double) Es 0 en el inicio de la recursión.
largo         : (int) Es un número entero fijo para todos los llamados a la recursión, y corresponde al largo del array original.

output:
suma          : (double) Suma de los números en el array usando el algoritmo "Pairwise Summation".

'''
def pairwise_summation(numbers, suma, largo):
    #Acá va su código
    if len(numbers) <= np.log2(largo):
      suma = numbers[0]
      for pos in range(1,len(numbers)):
        suma += numbers[pos]
    else:
      mitad = int(np.floor(len(numbers)/2))
      suma = pairwise_summation(numbers[0:mitad], suma, largo) + pairwise_summation(numbers[mitad:], suma, largo)

    return suma



# Firma función 7):
'''
input:
numbers       : (ndarray) Numpy Array 1D.

output:
suma          : (double) Suma de los números en el array usando el algoritmo "Kahan Summation".

'''
def kahan_summation(numbers):
    suma = 0
    varC = 0
    for i in range(len(numbers)):
      varY = numbers[i] - varC
      varT = suma + varY
      varC = (varT-suma) - varY
      suma = varT
    return suma

# Firma función 8):
'''
input:
numbers       : (ndarray) Numpy Array 1D.
real          : (double)  Número real con el cual se comparará cada resultado de las sumas.

output:
errors        : (ndarray) Array con los errores absolutos de cada algoritmo. Donde la posición 'i' del array representa al error
                          absoluto de la función 'i+1'. Por ejemplo el índice 0 representa el error absoluto de manual_summation 
                          con el número real esperado.

'''
def best_summation(numbers, real):
    largo = numbers.size
    resultados = np.array([manual_summation(numbers),
                    ascending_sort_summation(numbers),
                    descending_sort_summation(numbers),
                    random_summation(numbers),
                    numpy_summation(numbers),
                    pairwise_summation(numbers, 0, largo),
                    kahan_summation(numbers)])
    errors = np.abs(resultados-real)
    return errors


# Firma función 9):
'''
input:
K       : (int) cantidad de números que pertenecientes a la serie que se incluirán en el Numpy Array.

output:
numbers  : (ndarray) Numpy Array 1D con los números de la serie truncada hasta K términos.

HINTS:
'Maybe you wanna play with e_machine/2, that is the best way to get the value of K'
'''
'''
Usando los Hints se puede obtener la siguiente ecuación:
1   2^(-52)
- =  -----  -> K = 94.906.265
K^2   2
K = (2/(2**(-52)))**(1/2)
Básicamente queremos que el error que se va a cometer sea menor que e_machine/2
'''
def finding_terms(K):
    #Acá va su código
    numbers = np.empty(K)
    for i in range(0,K):
      numbers[i] = 1/((i+1)**2)

    return numbers
```
